```python
import requests

for i in range(250):
    url='https://movie.douban.com/j/new_search_subjects?sort=S&range=0,10&tags=%E7%94%B5%E5%BD%B1&start={}&genres=%E6%82%AC%E7%96%91'.format(i*20)
    headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.117 Safari/537.36"}
    response=requests.get(url,headers=headers).json()["data"]
    
    for movie in response:
        title=movie['title']
        rate=movie['rate']
        id=movie['id']
        director=movie['directors']
        
        print('{},{},{},{}'.format(title,rate,id,director))
    
```
